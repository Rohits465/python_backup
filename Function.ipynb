```python
def largest_number(x,y,z):
    if(x > y and x > z):
        print(x,"is largest")
    elif(y > x and y > z):
        print(y,"is largest")
    else:
        print(z,"is largest")
    
largest_number(5,6,7)
```

```python
def max_of_two(num_1,num_2):
    if (num_1 > num_2):
        return num_1
    else:
        return num_2

def max_of_three(num_1,num_2,num_3):
    return max_of_two(num_1,max_of_two(num_2,num_3))


result = max_of_three(5,6,7)
print(result,"is largest")
```

```python
# Q1 wap to print the fibbo. series using generator
# 0 1 1 2 3 5 8 13 21...

def fib(n):
    x = 0
    y = 1
    while(x<=n):
        yield x
        x,y = y,x+y

c = fib(13)
print(next(c))
print(next(c))
print(next(c))
```

```python
def mod_div(func):
    def xyz(a,b):
        if a < b:
            a,b =b,a
            return func(a,b)
        else:
            return func(a,b)
    return xyz

@mod_div
def div(x,y):
    z = x/y
    print(z)

div(2,4)
```

```python
def xyz(func):
    def pqr(*args,**kwargs):
        print("* "  * 15)
        func(*args,**kwargs)
        print("* " * 15 )
    return pqr

def abc(func):
    def pqr(*args,**kwargs):
        print("# " * 15)
        func(*args,**kwargs)
        print("# " * 15 )
    return pqr

# @xyz
# @abc
def show(msg):
    print(msg)
    
c = xyz(abc(show))
c("hello")
```
